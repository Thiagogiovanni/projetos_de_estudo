# Panorama do COVID-19

Segundo a **OMS** o coronavírus (COVID-19) é uma doença infecciosa causada pelo vírus SARS-CoV-2.

A maioria das pessoas infectadas com o vírus experimentará doenças respiratórias leves a moderadas e se recuperará sem a necessidade de tratamento especial. No entanto, alguns ficarão gravemente doentes e precisarão de atenção médica. Pessoas idosas e aqueles com condições médicas subjacentes, como doenças cardiovasculares, diabetes, doenças respiratórias crônicas ou câncer, são mais propensos a desenvolver doenças graves. Qualquer pessoa pode ficar doente com COVID-19 e ficar gravemente doente ou morrer em qualquer idade.

## **Dicionário das variáveis**

Deixarei aqui um resumo do dicionário das variáveis, porém o dicionário completo pode ser encontrado [aqui](https://github.com/owid/covid-19-data/tree/master/public/data)

* `total_cases	` - numéro total de casos de COVID-19
* `new_cases	` - novos casos 
* `new_cases_smoothed	` - novos casos nos últimos 7 dias
* `total_deaths` - número total de mortes
* `new_deaths` - novos casos de morte 
* `new_deaths_smoothed	` - novos casos de morte nos últimos 7 dias
* `hosp_patients` - número de pacientes hospitalizados no dia
* `total_tests` - total de testes de COVID-19
* `new_tests	` - novos testes de COVID-19
* `total_vaccinations` - informa o total de vacinas administradas
* `people_vaccinated` - total de pessoas que receberam ao menos uma dose da vacina
* `people_fully_vaccinated	` - total de pessoas que receberam todas as dose da vacina
* `total_boosters	` - total de vacinas de reforço administradas
* `new_vaccinations	` - novos casos de vacinação
* `new_vaccinations_smoothed` - novos casos de vacinação num período de 7 dias
* `continent	` - continente
* `location` - países
* `date` - data
* `population` - população do país em questão
* `population_density` - população dividido pelo "tamanho" do país
* `median_age	` - idade média
* `aged_65_older	` - quantidade de pessoas com mais de 65 anos
* `aged_70_older	` - quantidade de pessoas com mais de 70 anos
* `gdp_per_capita	` - representa o PIB/poder de compra no país
* `cardiovasc_death_rate` - número de mortes por doenças cardiovasculares (número anual por 100.000 pessoas)
* `diabetes_prevalence	` - porcentagem de diabéticos com idade entre 20 e 79 anos
* `female_smokers	` - quantidade de mulheres fumantes
* `male_smokers	` - quantidade de homens fumantes
* `handwashing_facilities		` - parcela da população que tem acesso as condições mínimas para lavar as mãos
* `life_expectancy` - expectativa de vida

## Países com mais casos e mortos por COVID.

Vamos entender quais países mais tiveram contaminados e mais mortos pelo COVID. Para isso vamos analisar sobre o dia atual (17/12/2022)

```python
df.loc[df.date == '2022-12-17']
```

```python
# vamos checar a quantidade de países para ver se bate com o tamanho do dataframa acima (236 linhas)

len(df.location.unique())
```

Aparecem um total de 248 países, porém para data de hoje temos informação apenas de 232.

Vamos fazer uma análise para entender como essa variável *location* está se comportando. Como queremos relacionar ela com o total de casos e o total de morte, faremos isso.


```python
df.loc[df.date == '2022-12-17', ['location','total_cases']].sort_values(by ='total_cases', ascending = False)[:10]
```

Veja que temos um conflito na variável *location*, pois a mesma tem valores como World e os continentes, como queremos apenas olhar para os países vamos tirar esses casos.

Para tirarmos esses casos de uma forma não tão manual (vendo 1 por 1) podemos notar que **aparentemente** quando aparece continentes ou World em *location* o valor de *continent* é Nan

Para tentar provar a hipótese acima faremos o seguinte:


```python
#vamos filtrar o dataframe e olhar apenas para os valores que possuem valores faltantes em continent
filter = df[df.continent.isnull()]
```

```python
#checando os valores de iso_code (paises possuem sua 'sigla' e regiões tem o prefixo OWID_)
filter.iso_code.unique()
```

Aqui vemos que de fato, todos valores de continente nulo não estão relacionados com países, mas com regiões e continentes.

Como queremos analisar os dados dos países podemos retirar esses casos.

```python
df.dropna(axis = 0, subset=['continent'], inplace = True)
```

Agora vamos olhar os 5 países com mais casos e mais mortes

```python
df.loc[df.date == '2022-12-17', ['location','total_cases']].sort_values(by ='total_cases', ascending = False)[:5]
```

Em relação ao total de casos, os países que lideravam até 17 de dezembro de 2022 eram:

* Estados Unidos - 99.887.614
* India - 44.677.175
* França - 38.841.776 


```python
df.loc[df.date == '2022-12-17', ['location','total_deaths']].sort_values(by ='total_deaths', ascending = False)[:5]
```

Em relação ao total de mortes, os países que lideravam até 17 de dezembro de 2022 eram:

* Estados Unidos - 1.087.410
* Brasil - 691.810
* India - 530.672

Vamos analisar os países com maior quantidade de mortes por COVID 19 em um gráfico de barras.

```python
#criando um dataframe apenas com os 5 países com maior número de mortes
top5_total_deaths = df.loc[df.date == '2022-12-17', ['location','total_deaths']].sort_values(by ='total_deaths', ascending = False)[:5]
```

```python
#fazendo o gráfico
fig, ax = plt.subplots(figsize = (12,8))

sns.barplot(x='location', y='total_deaths', palette="deep", ax=ax, data = top5_total_deaths)
```

Vamos analisar um gráfico de dispersão entre o PIB e a quantidade de mortos por milhão.

Se usarmos apenas o total de mortes teriamos algo inconclusível, pois estariamos analisando o valor absoluto, não o relativo. Ou seja, poderiamos ter, por exemplo, algum país com poucas pessoas mas que uma grande porcetagem morreu e se olharmos apenas para o total de mortes essa informação sumiria.

```python
#criando um dataframe com os parametros que usaremos
df_ultimo_dia = df.loc[df.date == '2022-12-17', ['location', 'gdp_per_capita','total_deaths_per_million']]
```

Marcarei uma reta vertical na média do PIB, para podermos analisar em volta desse valor.

```python
sns.set_theme(style="whitegrid")

# plotar o scatterplot
fig, ax = plt.subplots(figsize=(12,8))

#reta vertical
plt.axvline(x=df.gdp_per_capita.mean(), ymin=0, ymax=1)

#seta
plt.annotate('Média do PIB',
      xytext = (df.gdp_per_capita.mean()+25000,55),
      xy = (df.gdp_per_capita.mean(), 3),
      arrowprops = dict(facecolor = 'black', shrink = 0.05))

#dispersao
df_ultimo_dia.plot.scatter(x='gdp_per_capita', y='total_deaths_per_million', ax=ax, s=15)


plt.show()



```

```python
# analisando os países com maiores números de mortes por milhão

df_ultimo_dia.sort_values(by = 'total_deaths_per_million', ascending = False).head()
```

## Olhando para os Estados Unidos

Vamos criar um outro dataframe apenas com as informações sobre os EUA

```python
df_eua = df[df.location == "United States"].set_index('date')
```

Vendo graficamente como que o total de casos e total de morte se "comportaram" nos Estados Unidos.

```python
fig, ax = plt.subplots(nrows=2)
sns.set_theme(style="whitegrid")

# gráfico total de casos
df_eua["total_cases"].plot(kind="line", ax=ax[0], figsize = (12,7))
ax[0].set_title("Total de casos COVID nos EUA")
ax[0].set_ylabel("Casos Confirmados")
ax[0].set_xlabel("Data") 
ax[0].margins(x=0,y=0)
fig.autofmt_xdate()
plt.tight_layout()

# gráfico total de mortes
df_eua["total_deaths"].plot(kind="line", ax=ax[1], figsize = (12,7))
ax[1].set_title("Total de mortes por COVID nos EUA")
ax[1].set_ylabel("Mortes Confirmados")
ax[1].set_xlabel("Data") 
ax[1].margins(x=0,y=0)
fig.autofmt_xdate()
plt.tight_layout()
```

Podemos também analisar o gráfico de novos casos nos Estados Unidos

```python
fig, ax = plt.subplots(figsize = (12,8))

df_eua.new_cases.plot(ax=ax)
plt.xlabel('Data')
plt.ylabel('Total de casos')
plt.title('Novos casos nos EUA')
fig.autofmt_xdate()
plt.show()
```

Vamos verificar quando ocorreu o primeiro caso nos EUA e a primeira morte

```python
#primeiro caso
df_eua.loc[df_eua.total_cases != 0].head(1)
```

```python
#primeira morte
df_eua.loc[df_eua.total_deaths > 0].head(1)
```

Ou seja, temos que:

* Primeiro caso registrado foi no dia 22 de janeiro de 2020
* Primeira morte registrada foi no dia 29 de fevereiro de 2020

Vamos calcular quanto tempo se passou entre o primeiro caso e primeira morte


```python
from datetime import datetime

# Data primeiro caso
data_primeiro_caso = datetime.strptime('2020-01-22','%Y-%m-%d')

# Data primeira morte
data_primeira_morte = datetime.strptime('2020-02-29','%Y-%m-%d')

# Realizamos o calculo da quantidade de dias
quantidade_dias = abs((data_primeiro_caso - data_primeira_morte).days)
```

```python
print("Se passaram {} dias entre o primeiro caso e primeira morte".format(quantidade_dias))
```
